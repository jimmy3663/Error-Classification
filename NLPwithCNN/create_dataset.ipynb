## Application Failure Error (cate name: 3)

```python
from itertools import permutations
from string import ascii_lowercase

alphabet = list(ascii_lowercase)
len(alphabet)
```

```python
alpha_permute = list(permutations(alphabet,4))
print(alpha_permute[0])
```

```python
for i in range(150):
    name = ""
    for j in range(4):
        name += alpha_permute[i][j]
    e1 = "NameError: name '"
    e2 = "' is not defined"
    error = e1+name+e2
    error_dict[error] = 'Error from Applicatoin. Please check agsin'
```
